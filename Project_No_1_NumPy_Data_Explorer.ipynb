```python
import numpy as np
import time

print("===== NUMPY DATA EXPLORER =====\n")
```

```python
# 1 Array Creation
arr = np.array([10, 20, 30, 40, 50])
print("Original Array:", arr)
```

```python
# 2 Indexing & Slicing
print("\nIndexing:")
print("First element:", arr[0])
print("Last element:", arr[-1])
print("\nSlicing:")
print("First 3 elements:", arr[:3])
```

```python
# 3 Mathematical Operations
print("\nMathematical Operations:")
print("Add 5:", arr + 5)
print("Square:", arr ** 2)
```

```python
# 4 Axis-wise / 2D Operations
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])

print("\nMatrix:\n", matrix)
print("Sum of columns:", matrix.sum(axis=0))
print("Sum of rows:", matrix.sum(axis=1))
```

```python
# 5 Statistical Operations
print("\nStatistics:")
print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Standard Deviation:", np.std(arr))
```

```python
# 6 Reshaping
reshaped = np.arange(1, 10).reshape(3, 3)
print("\nReshaped Array (3x3):\n", reshaped)
```

```python
# 7 Broadcasting
print("\nBroadcasting Example:")
print(reshaped + 10)
```

```python
# 8 Save & Load Array
np.save("my_array.npy", arr)
loaded_arr = np.load("my_array.npy")
print("\nLoaded Array:", loaded_arr)
```

```python
# 9 Performance Comparison
print("\nPerformance Comparison:")

size = 1000000

list1 = list(range(size))
list2 = list(range(size))

start = time.time()
result = [x + y for x, y in zip(list1, list2)]
end = time.time()
print("Python List Time:", end - start)

arr1 = np.arange(size)
arr2 = np.arange(size)

start = time.time()
result_np = arr1 + arr2
end = time.time()
print("NumPy Time:", end - start)

print("\n===== END =====")
```
